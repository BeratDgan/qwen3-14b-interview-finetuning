```python
from unsloth import FastLanguageModel
import torch
model_name = "Qwen/Qwen3-14B"

max_seq_length = 4096  # Interview answers might be long
dtype = None
load_in_4bit = True    # For fitting 14B model to VRAM

# Qwen3-14B model
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = model_name,
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,

)

print(f"‚úÖ Model {model_name} loaded successfully. Ready for data preparation.")
print(f"üìä VRAM Kullanƒ±mƒ±: {torch.cuda.memory_allocated()/1024**3:.2f} GB")
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 64,  # We have A100 power, so use a high rank
    target_modules = [
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj"
    ],
    lora_alpha = 32,  # For r=64, alpha=32 is more balanced (commonly r/2 or r)
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth",  # Save VRAM
    random_state = 3407,
    use_rslora = False,
    loftq_config = None,
)

print("LoRA adapters have been added")
print(f"VRAM Usage: {torch.cuda.memory_allocated()/1024**3:.2f} GB")
```

```python
from datasets import load_dataset

# Define a standard interview-style prompt format (Alpaca-style)
alpaca_prompt = """Below is an interview question and the ideal answer a candidate should provide.

### Instruction:
{}

### Response:
{}"""

EOS_TOKEN = tokenizer.eos_token  # End-of-sequence token

def formatting_prompts_func(examples):
    # Adjust this section according to the column names in your JSONL file.
    # Commonly used keys are: "instruction" (question) and "output" (answer).
    # If your file uses different keys such as "question" and "answer",
    # replace them accordingly below.
    instructions = examples["instruction"]
    outputs      = examples["output"]
    texts = []
    for instruction, output in zip(instructions, outputs):
        # Combine the interviewer-style prompt with the expected response
        text = alpaca_prompt.format(instruction, output) + EOS_TOKEN
        texts.append(text)
    return { "text": texts }

# Load the dataset (adjust the file name if needed)
dataset = load_dataset("json", data_files="/content/drive/MyDrive/Colab Notebooks/final_train_ready.jsonl", split="train")
dataset = dataset.map(formatting_prompts_func, batched=True)

print(f"Total Number of Samples: {len(dataset)}")
print("Sample Data:", dataset[0]["text"])
```

```python
# ==============================================================
# CELL 5: TRAINING ‚Äì A100 OPTIMIZED
# ==============================================================

from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    dataset_num_proc = 4,  # 4 workers perform better on A100
    packing = False,       # More stable for long sequences
    args = TrainingArguments(
        # === BATCH SIZE ===
        per_device_train_batch_size = 4,   # 4 instead of 8 ‚Äì more stable
        gradient_accumulation_steps = 4,   # Effective batch = 4 * 4 = 16

        # === TRAINING DURATION ===
        num_train_epochs = 1,              # 1 epoch is sufficient for ~10K samples
        # max_steps = 1000,                # Alternative: step-based training

        # === LEARNING RATE ===
        learning_rate = 2e-4,
        warmup_ratio = 0.03,               # Ratio instead of warmup_steps (more flexible)
        lr_scheduler_type = "cosine",      # Cosine instead of linear (better)

        # === PRECISION ===
        fp16 = not is_bfloat16_supported(),
        bf16 = is_bfloat16_supported(),    # Use bf16 on A100

        # === OPTIMIZATION ===
        optim = "adamw_8bit",
        weight_decay = 0.01,

        # === LOGGING & SAVING ===
        logging_steps = 25,                # More frequent logging (for 10K samples)
        save_strategy = "steps",
        save_steps = 500,                  # Checkpoint every 500 steps
        save_total_limit = 3,              # Keep max 3 checkpoints

        # === MISC ===
        seed = 3407,
        output_dir = "outputs",
        report_to = "none",                # If not using wandb
    ),
)

# GPU memory status
gpu_stats = torch.cuda.get_device_properties(0)
print(f"GPU:  {gpu_stats.name}")
print(f"VRAM: {gpu_stats.total_memory / 1024**3:.2f} GB")
print(f"Used:  {torch.cuda.memory_allocated() / 1024**3:.2f} GB")
print("Training start...")
```

```python
# START TRAINING
trainer_stats = trainer.train()
```

```python
import os
import shutil

# 1. Define paths
# Local path (Fast SSD inside Colab)
local_model_path = "Qwen3-14B-Interview-GGUF"
# Drive path (Destination)
drive_folder = "/content/drive/MyDrive/Colab_Models/Qwen3-14B-Interview"

# Create Drive folder if not exists
if not os.path.exists(drive_folder):
    os.makedirs(drive_folder)

print("Starting GGUF Conversion on Local Disk (Stable & Fast)...")

# 2. Save GGUF locally first to avoid Drive Timeout/Empty File errors
model.save_pretrained_gguf(
    local_model_path,
    tokenizer,
    quantization_method = "q4_k_m",
)

print("Conversion finished locally!")
print(f"Moving the GGUF file to Google Drive: {drive_folder} ...")

# 3. Move the finished file to Drive safely
# We look for the .gguf file in the local folder
for file in os.listdir(local_model_path):
    if file.endswith(".gguf"):
        source = os.path.join(local_model_path, file)
        destination = os.path.join(drive_folder, file)

        # Copy file
        shutil.copy(source, destination)
        print(f"Successfully saved to Drive: {destination}")

print("ALL DONE! You can close the tab now.")
```
